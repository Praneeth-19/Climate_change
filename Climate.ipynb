```python
land_avg['Year'] = land_avg['dt'].dt.year
average_per_year = land_avg.groupby('Year')[['LandAverageTemperature', 'LandAverageTemperatureUncertainty']].mean()
average_per_year
```

```python
average_per_year['lower temp'] = average_per_year['LandAverageTemperature']-average_per_year['LandAverageTemperatureUncertainty']
average_per_year['upper temp'] = average_per_year['LandAverageTemperature']+average_per_year['LandAverageTemperatureUncertainty']

plt.figure(figsize=(16,6))
plt.plot(average_per_year.index, average_per_year['LandAverageTemperature'], color='red', label='Average')
plt.plot(average_per_year.index, average_per_year['upper temp'], color='blue', alpha=0.3)
plt.plot(average_per_year.index, average_per_year['lower temp'], color='blue', alpha=0.3)
plt.fill_between(average_per_year.index, average_per_year['upper temp'], average_per_year['lower temp'], color='lightblue', alpha=0.3, label='Temperature Uncertaintry boundaries')
plt.title('Average Temperature per Year')
plt.legend(loc='best')
```

```python
from statsmodels.tsa.api import VECM
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_percentage_error

df['Year'] = df['dt'].dt.year

df_avg = df.groupby('Year')[['LandAverageTemperature', 'LandAndOceanAverageTemperatureUncertainty']].mean()[100:]

scaler = StandardScaler()
data_sc = pd.DataFrame(scaler.fit_transform(df_avg[['LandAverageTemperature', 'LandAndOceanAverageTemperatureUncertainty']]), columns=['LandAverageTemperature', 'LandAndOceanAverageTemperatureUncertainty'])
data_sc.index = df_avg.index
train = data_sc.iloc[:-20]

test = pd.DataFrame(scaler.inverse_transform(data_sc.iloc[-20:]), columns=data_sc.columns)
test.index = data_sc.iloc[-20:].index

    
vecm_model = VECM(endog=train, k_ar_diff=1, coint_rank=0, deterministic='co')
vecm_fit = vecm_model.fit()
pred = pd.DataFrame(scaler.inverse_transform(pd.DataFrame(vecm_fit.predict(steps=20), columns=['LandAverageTemperature', 'LandAndOceanAverageTemperatureUncertainty'])), columns=['LandAverageTemperature', 'LandAndOceanAverageTemperatureUncertainty'])
pred.index = test.index

train = pd.DataFrame(scaler.inverse_transform(train), columns=['LandAverageTemperature', 'LandAndOceanAverageTemperatureUncertainty'])
train.index = data_sc.iloc[:-20].index
        
mape = mean_absolute_percentage_error(test['LandAverageTemperature'], pred['LandAverageTemperature'])
print(mape)
plt.figure(figsize=(16,6))
plt.plot(train.index, train['LandAverageTemperature'], label='train')
plt.plot(test.index, test['LandAverageTemperature'], label='test')
plt.plot(pred.index, pred['LandAverageTemperature'], label='pred')
plt.xlabel('LandAverageTemperature')
plt.legend(loc='best')
```
